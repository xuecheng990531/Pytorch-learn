```python
import torch,torchvision
from torch.utils.data import DataLoader
from torch.nn import Module,Conv2d,MaxPool2d,Flatten,Linear,Sequential
from Mymodel import *
from torch.utils.tensorboard import SummaryWriter
```

```python
train_data=torchvision.datasets.CIFAR10(
    root='/Users/lixuecheng/Desktop/Pytorch learning copy/Dataset',train=True,
    transform=torchvision.transforms.ToTensor(),download=True)

test_data=torchvision.datasets.CIFAR10(
    root='/Users/lixuecheng/Desktop/Pytorch learning copy/Dataset',train=False,
    transform=torchvision.transforms.ToTensor(),download=True)
```

```python
train_data_loader=DataLoader(train_data,batch_size=64)
test_data_loader=DataLoader(test_data,batch_size=64)
```

```python
learning_rate=0.01
tudou=Mymodel()
loss=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(tudou.parameters(),lr=learning_rate)
writer=SummaryWriter('logs_test')
```

```python
total_test_step=0
total_train_step=0
total_acc=0
total_loss=0

for i in range(5):
    print("---------第 {} 轮训练---------".format(i+1))

    for data in train_data_loader:
        images,targets=data
        output=tudou(images)
        loss_result=loss(output,targets)

        optimizer.zero_grad()
        loss_result.backward()
        optimizer.step()

        total_train_step=total_train_step+1

        accuracy=(output.argmax(1)==targets).sum()

        total_acc=total_acc+accuracy

        real_acc=(total_acc/len(train_data))*100

        writer.add_scalar('train_data_loss',loss_result,total_train_step)

        print('准确率 {}'.format((total_acc/len(train_data))*100)+'%')

    
    with torch.no_grad():
        for data in test_data_loader:
            imaegs,targets=data
            output=tudou(images)
            loss_result=loss(output,targets)

            total_loss=total_loss+loss

            total_test_step=total_test_step+1

            writer.add_scalar('test_data',total_loss,total_test_step)

            print('Loss: {}'.format(total_loss))

writer.close()
```
