```python
 import torch
 import torchvision
 from torch.utils.data import DataLoader
 import torch.nn as nn
 import torch.nn.functional as f
 from torch.utils.tensorboard import SummaryWriter
```

```python
dataset=torchvision.datasets.CIFAR10(
    root='/Users/lixuecheng/Desktop/Pytorch learning copy/Dataset',
    train=False,transform=torchvision.transforms.ToTensor(),download=True
)
```

```python
dataloader=DataLoader(dataset=dataset,batch_size=64)
```

```python
class Mymodel(nn.Module):
    def __init__(self):
        super(Mymodel,self).__init__()
        self.conv1=nn.Conv2d(
            in_channels=3,
            out_channels=6,
            kernel_size=3,
            padding=0
        )


    def forward(self,x):
        x=f.relu(self.conv1(x))
        return x
```

```python
cifar10=Mymodel()
writer=SummaryWriter('cifar10')
```

```python
step=0
for i,(images,targets) in enumerate(dataloader):
    output=cifar10(images)
    print("images_shape:",images.shape) 
    print("out_put shape:",output.shape)

    
    writer.add_images('inputs',images,step)


    # 需要对输出的图像形状进行更改，-1表示自己计算大小，不需要人为进行指定
    output=torch.reshape(output,(-1,3,30,30))


    writer.add_images('outputs',output,step)    
    step=step+1
    
```
