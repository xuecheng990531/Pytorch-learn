```python
import torch
from torch.utils.tensorboard import SummaryWriter
from torch.utils.data import DataLoader
import torchvision
import torch.nn as nn
```

```python
dataset=torchvision.datasets.CIFAR10(
    root='/Users/lixuecheng/Desktop/Pytorch learning copy/Dataset',
    train=False,transform=torchvision.transforms.ToTensor(),download=True
)
dataset
```

```python
dataloader=DataLoader(dataset=dataset,batch_size=16,drop_last=True)
```

```python
writer=SummaryWriter('log_liner')
```

```python
class Mymodel(nn.Module):
    def __init__(self):
        super(Mymodel,self).__init__()
        self.liner1=nn.Linear(49152,10)
    def forward(self,input):
        output=self.liner1(input)
        
        return output
```

```python
tudou=Mymodel()
```

```python
for i in dataloader:
    images,targets=i
    print(images.shape)

    output=torch.flatten(images)
    print(output.shape)
    
    output2=tudou(output)
    print(output2.shape)
```
